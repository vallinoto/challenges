```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor, DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
import sklearn.metrics as metrics
from pandas_profiling import ProfileReport

pd.set_option("display.max_columns", 200)

FILE_TRAIN = '../data/train.csv'
FILE_TEST = '../data/test.csv'
```

```python
enem_train = pd.read_csv(FILE_TRAIN, sep=',', encoding='utf-8')
enem_test = pd.read_csv(FILE_TEST, sep=',', encoding='utf-8')
```

```python
 profile = ProfileReport(enem_train)
```

```python
features_selecionadas = correlacao_features[(correlacao_features['NU_NOTA_MT'] <= -0.28) | 
                       (correlacao_features['NU_NOTA_MT'] >= 0.28) & 
                       (correlacao_features['NU_NOTA_MT'] < 1.0)]['NU_NOTA_MT']
features_selecionadas = features_selecionadas.index.to_list()
features_selecionadas
```

```python
enem_train[features_selecionadas].isnull().sum()
```

```python
for feature in features_selecionadas:
        enem_train[feature].fillna(0, inplace=True)
        enem_test[feature].fillna(0, inplace=True)
```

```python
enem_train['NU_NOTA_MT'].isnull().sum()
```

```python
enem_train['NU_NOTA_MT'].fillna(0, inplace=True)
```

```python
enem_train[features_selecionadas].isnull().sum()
```

```python
enem_test[features_selecionadas].isnull().sum()
```

```python
enem_train['NU_NOTA_MT'].isnull().sum()
```

```python
y_train = enem_train['NU_NOTA_MT']
X_train = enem_train[features_selecionadas]
X_test = enem_test[features_selecionadas]
```

```python
scaler = StandardScaler()  
X_train = scaler.fit_transform(X_train)  
X_test = scaler.transform(X_test)
```

```python
mdl = RandomForestRegressor( 
           criterion='mae', 
           max_depth=8,
           max_leaf_nodes=None,
           min_impurity_split=None,
           min_samples_leaf=1,
           min_samples_split=2,
           min_weight_fraction_leaf=0.0,
           n_estimators= 500,
           n_jobs=-1,
           random_state=0,
           verbose=0,
           warm_start=False
)
```

```python
mdl.fit(X_train, y_train)
```

```python
predicao = mdl.predict(X_train)
```

```python
print('MAE:', metrics.mean_absolute_error(y_train, predicao).round(8)  )
print('MSE:', metrics.mean_squared_error(y_train, predicao).round(8) )  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train, predicao)).round(8))
```

```python
predicao = mdl.predict(X_test)
```
